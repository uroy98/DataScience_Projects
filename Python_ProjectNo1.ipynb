### 4. Write a program to calculate the area of a circle. The program should ask for an appropriate input and print the computed area. 
### a. Input text: Please enter the radius of the circle.
### b. If the user enters a value of 5. The output should be: The area of the circle is 78.54
### c. This program should indicate that the input is invalid if the user enters a character instead of a number as input. For instance if the user enters a value of ‘a’ instead of 5 in the previous example the program should prompt: Invalid input, please enter a number: 

```python
import math
r = input("Please enter the radius of the circle: ") #taking input radius of the circle
while True:
    if r.isnumeric() == False: #checking if input is a non-integer value or not
        print("Invalid input, please enter a number: ")
        r = input("Please enter the radius of the circle: ") #taking correct value once again if previous input is incorrect
    elif r.isnumeric() == True:
        break #breaking from loop when input is a number
r = int(r)
area = math.pi*r*r
print("The area of the circle is %.2f" %area)
```

### 5. Write a program which will take two lists X & Y as input and it will return a list having all the odd numbers present in the two lists. (Note:- common elements can be repeated) 
### Input list: X=[1,2,3,4,5] & Y=[5,6,7,8,9]
### Expected Output: [1,3,5,5,7,9] or [1,3,5,7,9] both are accepted.

```python
X=[1,2,3,4,5] #input list X
Y=[5,6,7,8,9] #input list Y
L = []
FinalList = []
for x in X:
    if x%2!=0:
        L.append(x) #odd numbers from list X is appended in list L
for y in Y:
    if y%2!=0:
        L.append(y) #odd numbers from list Y is appended in list L after inserting odd numbers from list X
for x in L:
    if x not in FinalList:
        FinalList.append(x) #keeping only the unique odd numbers in the final list FinalList
print(FinalList)
```

### 6. Write a function that prints the integers from 1 to 15. But for multiples of three print "GREAT" instead of the number, and for the multiples of five print "LEARNING". For numbers which are multiples of both three and five print "GREATLEARNING".

```python
L = list(range(1,16)) #input list ranging from integers 1 to 15
for x in L:
    if x%3==0 and x%5!=0:
        index = L.index(x)
        L[index] = "GREAT" #For multiples of only three print "GREAT" instead of the number
    elif x%3!=0 and x%5==0:
        index = L.index(x)
        L[index] = "LEARNING" #For multiples of only five print "LEARNING" instead of the number
    elif x%3==0 and x%5==0:
        index = L.index(x)
        L[index] = "GREATLEARNING" #For multiples of both three and five print "GREATLEARNING" instead of the number
print(L)
```
